### List of HS codes from UN

Preprocess and clean list

```python
HS = pd.read_csv("Data/HS_commodities.csv")
HS = HS[HS['Level'] == 4].reset_index(drop = True)
list_HS = comtrade.hs_list(HS)
```

Searching for HS codes depending on a keyword

```python
keyword = ['cars']

codes  = comtrade.keyword_search(list_HS, keyword)
display(codes.style.set_properties(**{'white-space': 'pre-wrap',}))
```
